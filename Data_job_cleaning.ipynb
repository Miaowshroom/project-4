```python
import pandas as pd
```

```python
job_post = pd.read_csv('D:\GA\project_submission\project-4\datajob\datajob\data_all_100pages.csv', na_filter=False)
```

```python
job_title = job_post.job_title
```

```python
bag_of_word_title = ['data', 'machine learning', 'analyst', 'analytics', 'analysis', 'research', 'database', 'security','systems engineer','consultant','algorithms' ]
bag_of_word_role = ['analytics', 'data analysis', 'database', 'data security', 'data center' ]
bag_of_word_skill = ['machine learning','deep learning', 'business intelligence','computer vision', 'database', 'modeling', 'databases', 'recognition','detection', 'risk', 'mining', 'algorithms']
```

```python
check_job = 0
remove_list = []
title_matched = []
role_matched = []
skill_matched = []
for i in range(len(job_post)):
    
    find_title = False
    find_role = False
    find_skill = False
    
    for word in bag_of_word_title:
        if word in job_post.loc[i,'job_title'].lower():
            find_title = True
            title_matched.append(i)
            break
    
    if not find_title:
        for word in bag_of_word_role:
            if word in job_post.loc[i, 'job_role_resp'].lower():
                find_role = True
                role_matched.append(i)
                break
                
    if not(find_title | find_role):
        for word in bag_of_word_skill:
            if word in job_post.loc[i, 'job_skills'].lower():
                find_skill = True
                skill_matched.append(i)
                break
        
    if not(find_title | find_role | find_skill):
        print(f"For job {job_post.loc[i,'job_title']} \nthe required job skill are \n{job_post.loc[i,'job_skills']}\n\n")
        check_job +=1
        remove_list.append(i)
```

```python
keep_list = [i for i in range(len(job_post)) if i not in remove_list]
```
